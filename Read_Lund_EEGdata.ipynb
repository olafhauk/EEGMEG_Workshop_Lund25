# Reading Lund EEG data

```python
import mne
```

## 64-channel data

```python
fname = 'I:/users/olaf/Methods/EEGMEG_Workshops_Olaf/Lund_Apr25/EEG_data/EEG_transprobs_64ch/138_transprobs_preprocessed.fif'
raw = mne.io.read_raw_fif(fname)
```

```python
events, event_id = mne.events_from_annotations(raw, event_id=int)
epochs = mne.Epochs(raw, events, event_id=[10], tmin=-0.2, tmax=0.5)
evoked = epochs.average().set_eeg_reference()
evoked.plot_joint();
```
